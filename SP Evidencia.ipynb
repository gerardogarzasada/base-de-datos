```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
```

```python
df=pd.read_csv('datasets/headbrain.csv')
df.head(2)
```

```python
Head_size = df['Head Size(cm^3)']
Gender = df['Gender']
Brain_weight = df['Brain Weight(grams)']
```

**Pregunta 1**

¿Cómo es la distribución del BrainWeight?

```python
sns.boxplot(Brain_weight ).set_title("Diagrama de cajas y bigotes de Brain Weight")
plt.show()
```

```python
media = Brain_weight.mean()
mediana= Brain_weight.median()
mínimo= Brain_weight.min()
máximo= Brain_weight.max()

print ("La media es:",media)
print("La mediana es:",mediana)
print("El mínimo es:",mínimo)
print("El máximo:",máximo)
```

**Pregunta 2**

¿Cómo es la distribución de los datos de Head size?

```python
df['Head Size(cm^3)'].plot(kind='hist').set_title("Histograma de head size")
plt.show()
```

Simétrica centrada porque los valores de las variables ocurren a frecuencias regulares y cuando la media, la mediana y la moda ocurren en el mismo punto

Pregunta 3

¿Cómo es la relación entre la variable Head Size(cm^3) y Brain Weight(grams)?

```python
sns.lmplot(
    x='Head Size(cm^3)',
    y= 'Brain Weight(grams)',
    hue= 'Gender',
    fit_reg= True, 
    data = df
)
plt.show()
```

Pregunta 3

¿Con que ecuación se puede hacer una estimación del peso del cerebro?

```python
model= sm.OLS.from_formula('Brain_weight~Gender+Head_size', 
                          data=df)
result=model.fit()
result.summary()
```

ecuación:

BrainWeight= 396.5754 + 0.2509(Head Size) -17.8615(Gender)
